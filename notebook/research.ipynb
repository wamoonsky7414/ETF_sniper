```python
import pandas as pd
from pathlib import Path

# 設定資料夾路徑
folder_path = Path(r'C:\Users\User\Documents\GitHub\ETF_sniper\data\00981A')

# 取得目錄下所有 parquet 檔案
files = folder_path.glob('*.parquet')

df_list = []

for file in files:
    # 讀取 parquet 檔案
    temp_df = pd.read_parquet(file)
    
    # 取得檔名 (不含副檔名)，例如 '20251219'
    date_str = file.stem
    
    # 新增日期欄位，並將其轉換為 datetime 格式 (選擇性)
    temp_df['Date'] = pd.to_datetime(date_str)
    
    df_list.append(temp_df)

# 合併所有 DataFrame
if df_list:
    final_df = pd.concat(df_list, ignore_index=True)
    print(final_df.head())
else:
    print("找不到任何 Parquet 檔案。")
```
